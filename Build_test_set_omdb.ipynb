```python
import pandas as pd
import numpy as np

import requests 
from pprint import pprint

Cont_2019_db = pd.read_csv('./DataSources/theContendersDF_OMBD.csv')
Cont_2019_db
```

```python
# use api.tmdb.org due to problem of 'query limit is 40'
base_url ='https://api.tmdb.org/3/search/movie?'
api_key = 'api_key=d5b4e5d5145be601d4640d76efd791e3'
query = '&query='
query_base_url = base_url + api_key + query

# best_pic_info: get the movie summary information 
best_pic_2019 = []

for i in range(len(Cont_2019_db_name)): 
    release_year = '2018'
    movie_info = query_base_url + Cont_2019_db_name[i] + '&year=' + release_year
    print(movie_info)
    response=requests.get(movie_info).json()
    best_pic_2019.append(response['results'])
 
```

```python
base_url1 = 'https://api.tmdb.org/3/movie/'

#language = '&language=en-US'
keyword = '/keywords?'
Winner2019_Details= []
keyword_details = []
tmdb_id_list = []
#keywords: https://api.themoviedb.org/3/movie/23383/keywords?api_key=d5b4e5d5145be601d4640d76efd791e3

for i in range(len(best_pic_2019)):
    tmdb_id = str(best_pic_2019[i][0]['id'])
    tmdb_id_list.append(best_pic_2019[i][0]['id'])
    detail_query_link = base_url1 + tmdb_id + '?' + api_key #+ language
    keywords_query_link = base_url1 + tmdb_id + keyword + api_key
    response = requests.get(detail_query_link).json()
    keywords_response = requests.get(keywords_query_link).json()
    Winner2019_Details.append(response)
    keyword_details.append(keywords_response)
```

```python
winning_keywords = []
keywords_list = []

for i in range(len(keyword_details)):
    for j in range(len(keyword_details[i]['keywords'])):
        winning_keywords.append(keyword_details[i]['keywords'][j]['name'])
    keywords_list.append(keyword_details[i]['keywords'])

```

```python
revenue_list = []
budget_list = []
runtime_list = []
vote_average_list = []
vote_count_list =[]
pop_list = []
genre_list =[]
release_date_list = []


for i in range(len(Winner2019_Details)):
    revenue_list.append(Winner2019_Details[i]['revenue'])
    budget_list.append(Winner2019_Details[i]['budget'])
    runtime_list.append(Winner2019_Details[i]['runtime'])
    vote_average_list.append(Winner2019_Details[i]['vote_average'])
    vote_count_list.append(Winner2019_Details[i]['vote_count'])
    pop_list.append(Winner2019_Details[i]['popularity'])
    genre_list.append(Winner2019_Details[i]['genres'])
    release_date_list.append(Winner2019_Details[i]['release_date'])


release_month_list = []
for i in range(len(release_date_list)):
    release_month = release_date_list[i].split('-')
    release_month_list.append(release_month[1])
    

Cont_2019_db['id'] = pd.Series(tmdb_id_list).values
Cont_2019_db['revenue'] = pd.Series(revenue_list).values
Cont_2019_db['budget'] =  pd.Series(budget_list).values
Cont_2019_db['runtime'] = pd.Series(runtime_list).values
Cont_2019_db['vote_average'] = pd.Series(vote_average_list).values
Cont_2019_db['vote_count'] = pd.Series(vote_count_list).values
Cont_2019_db['popularity'] = pd.Series(pop_list).values
Cont_2019_db['genres'] = pd.Series(genre_list).values
Cont_2019_db['release_month'] = pd.Series(release_month_list).values
Cont_2019_db['keywords'] = pd.Series(keywords_list).values
```

```python
def get_names(keywords):
    return '|'.join([x['name'] for x in keywords]) 



Cont_2019_db['keywords'] = Cont_2019_db['keywords'].apply(get_names)
Cont_2019_db['genres'] = Cont_2019_db['genres'].apply(get_names)

Cont_2019_db.head()
```

```python
genre_weight_dict = {'Drama': 0.35911602209944754,\
                     'Romance': 0.12154696132596685,\
                     'History': 0.09392265193370165,\
                     'Comedy': 0.09392265193370165,\
                     'Crime': 0.06077348066298342,\
                     'War': 0.06077348066298342,\
                     'Thriller': 0.049723756906077346,\
                     'Music': 0.04419889502762431,\
                     'Adventure': 0.03867403314917127,\
                     'Action': 0.03314917127071823,\
                     'Family': 0.016574585635359115,\
                     'Western': 0.011049723756906077,\
                     'Fantasy': 0.011049723756906077,\
                     'Mystery': 0.0055248618784530384,\
                     'other': 0.07142857142857141}
```

```python
genre_list = list(Cont_2019_db['genres'])
w_genre_list = genre_weight_dict.keys()

weight=0
weight_list = []

for i in range(len(genre_list)):
    onefilm = genre_list[i].split('|')
    for j in range(len(onefilm)):
        if onefilm[j] in w_genre_list:
            weight  += genre_weight_dict[onefilm[j]]
        else:
             weight  += genre_weight_dict['other']
        
    weight_list.append(weight)
    weight = 0
```

```python
Cont_2019_db['genre_weight'] = weight_list
```

```python
Cont_2019_db.release_month = Cont_2019_db.release_month.astype(int)
```

```python
Cont_2019_db['profit'] = Cont_2019_db['revenue'] - Cont_2019_db['budget']
```
